```python
import numpy as np

data=np.random.normal(loc=100,scale=10,size=100)
sample=data[np.random.choice(len(data),size=30,replace=False)]

U=np.std(sample,ddof=1)
print(f"不偏標準偏差 U:{U:.4f}")
```

```python
S_avg=np.mean(sample)
Y=U/np.sqrt(len(sample))
div=S_avg-np.mean(data)
T=div/Y
print(T)
```

```python
num_samples=5000
T_values = []
#appendを使うために、あらかじめ空のリストを用意
for _ in range(num_samples):
  sample=data[np.random.choice(len(data),size=30,replace=False)]
  S_avg=np.mean(sample)
  Y=U/np.sqrt(len(sample))
  div=S_avg-np.mean(data)
  T=div/Y
  T_values.append(T)
  #appendはリストにしか使えないため、計算されたTの値をリストに追加

import matplotlib.pyplot as plt
plt.hist(T_values,bins=100,density=True) # Use the list of T values for plotting
plt.show()
```

```python
from scipy import stats

n=len(sample)
df=n-1
confidence_level=0.95
alpha=1-confidence_level
T_value_right=stats.t.ppf(1-alpha/2,df)
T_value_left=stats.t.ppf(alpha/2,df)

pop_avg_right=S_avg-T_value_left
pop_avg_left=S_avg-T_value_right

print(T_value_right)
print(f"95%の信頼区間: [{pop_avg_left:.3f},{pop_avg_right:.3f}]")
```
